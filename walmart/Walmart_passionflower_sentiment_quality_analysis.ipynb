```python
import pandas as pd
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#file Path
file_path = "c:/Drive_d/jobs/CrowdDoing/walmart/Walmart_Passion_flower_reviews.csv"

#count for encoding
df = pd.read_csv(file_path, encoding='ISO-8859-1')

df = df.drop(columns=["Unnamed: 3"])

df['product_id'] = df['product_id'].fillna(0).astype(int)

df['Rating'] = df['Rating'].apply(lambda x: '/'.join(x.split('/')[::-1]))
df = df.drop_duplicates(subset=['Product_name', 'product_id', 'User', 'Review_title', 'Review_description', 'Rating'])
```

```python
df
```

```python
# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Combine 'Review_title', 'Review_description' and 'Rating' into a single text, replacing NaN with an empty string
df['combined_review'] = df['Review_title'].fillna('') + " " + df['Review_description'].fillna('')

# Apply sentiment analysis
df['sentiment_scores'] = df['combined_review'].apply(lambda x: sia.polarity_scores(x))

# Extract sentiment classification based on compound score
df['sentiment'] = df['sentiment_scores'].apply(lambda x: 'positive' if x['compound'] > 0 else 'negative' if x['compound'] < 0 else 'neutral')


df
```

```python
# 1. Check for Missing Values
missing_values = df.isnull().sum()

# 2. Check for Data Consistency in Rating

df['Rating_parts'] = df['Rating'].apply(lambda x: x.split('/'))

# 3. Check for Duplicates in Columns (excluding non-hashable Rating_parts)
duplicates = df[['Product_name', 'product_id', 'User', 'Review_title', 'Review_description', 'Rating']].duplicated().sum()

# 4. Distribution of Ratings
rating_counts = df['Rating'].value_counts()

# 5. Check for Empty or Meaningless Reviews
df['Review_length'] = df['Review_description'].apply(lambda x: len(str(x).split()))  # Safely convert to string before split
empty_reviews = df[df['Review_length'] == 0]  # Reviews with 0 words

# 6. Integrity of Relationships (this would require an external data check, for example, product_id validation)

# Summary of quality checks
quality_summary = {
    "Missing Values": missing_values,
    "Duplicates": duplicates,
   "Rating Distribution": rating_counts,
    "Empty Reviews": empty_reviews,
}

# Display results
for key, value in quality_summary.items():
    print(f"{key}:")
    print(value)
    print("\n")
```

```python
df_passionflower = df.drop(columns=["Rating_parts","Review_length"])

df_passionflower
```
